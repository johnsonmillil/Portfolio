This Jupyter Notebook forecasts daily hospital revenue (`Revenue`) using ARIMA on `medical_clean.csv` for Horizon Health Network, addressing the research question: “Can we forecast daily hospital revenue over the next three months to optimize resource allocation?” (Scenario 1).

```python
# Import required libraries
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from scipy.signal import periodogram

# Note: 'seaborn' style removed due to installation issue; using default matplotlib style
```

## D1: Data Preparation and Visualization
Load `medical_clean.csv`, assign dates (May 19, 2023, to May 18, 2025), and create a line graph of daily revenue.

```python
# Load the dataset
df = pd.read_csv('medical_clean.csv')
print("Dataset loaded:")
print(df.head())
print(f"Number of rows in df: {len(df)}")

# Assign dates (May 19, 2023, to May 18, 2025)
df['Date'] = pd.date_range(start='2023-05-19', periods=len(df), freq='D')
df.set_index('Date', inplace=True)
df.index.freq = 'D'
print("\nFirst 5 and last 5 dates in df:")
print(df.index[:5])
print(df.index[-5:])

# Display the prepared data
print("\nDaily time series (Revenue in millions):")
print(df.head())
print("\nLast 10 days of time series:")
print(df.tail(10))
print(f"Number of days in df: {len(df)}")

# Save dataset
df.to_csv('medical_time_series.csv')
print("Time series saved as medical_time_series.csv")

# Visualize time series
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Revenue'], label='Daily Revenue')
plt.title('Daily Hospital Revenue Time Series')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions of Dollars)')
plt.legend()
plt.grid(True)
plt.savefig('time_series_plot.png')
plt.show()
```

## D2-D3: Time Step and Stationarity
The time series is daily (731 days, May 19, 2023, to May 18, 2025). Check stationarity with ADF test.

**Output**: ADF Statistic: -2.218, p-value: 0.200 (Non-stationary). After differencing: ADF Statistic: -17.375, p-value: 5.11e-30 (Stationary).

```python
# Stationarity test
result = adfuller(df['Revenue'].dropna())
print("\nADF Test Results:")
print(f"ADF Statistic: {result[0]:.3f}")
print(f"p-value: {result[1]:.3f}")
print("Stationary" if result[1] < 0.05 else "Non-stationary")

# Difference if non-stationary
if result[1] >= 0.05:
    df['Revenue_diff'] = df['Revenue'].diff().dropna()
    print("\nDifferenced time series:")
    print(df['Revenue_diff'].head())
    result_diff = adfuller(df['Revenue_diff'].dropna())
    print("\nADF Test on Differenced Data:")
    print(f"ADF Statistic: {result_diff[0]:.3f}")
    print(f"p-value: {result_diff[1]:.3f}")
    print("Stationary" if result_diff[1] < 0.05 else "Non-stationary")
else:
    df['Revenue_diff'] = df['Revenue']
```

## E1: Time Series Analysis and Visualizations
Analyze autocorrelation, spectral density, decomposition, and residuals.

```python
# Autocorrelation and Partial Autocorrelation
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(df['Revenue'].dropna(), lags=30, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')
plt.subplot(122)
plot_pacf(df['Revenue'].dropna(), lags=30, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF)')
plt.savefig('acf_pacf_plot.png')
plt.show()

# Spectral density
f, Pxx = periodogram(df['Revenue'])
plt.figure(figsize=(10, 6))
plt.semilogy(f, Pxx)
plt.title('Spectral Density')
plt.xlabel('Frequency')
plt.ylabel('Power Spectral Density')
plt.savefig('spectral_plot.png')
plt.show()

# Decomposition
decomposition = seasonal_decompose(df['Revenue'], model='additive', period=7)
decomposition.plot()
plt.savefig('decomposition_plot.png')
plt.show()

# Residuals
residuals = decomposition.resid
plt.figure(figsize=(10, 6))
plt.plot(residuals, label='Residuals')
plt.title('Residuals of Decomposed Time Series')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend()
plt.grid(True)
plt.savefig('residuals_plot.png')
plt.show()
```

## E2-E3: ARIMA Modeling and Forecasting
Fit ARIMA(1,1,1) and forecast 90 days.

**Output**:
- Train size: 584, Test size: 147
- ARIMA(1,1,1) Summary: AIC=707.136, AR(1) coef=0.4596 (p=0.000), MA(1) coef=-0.0587 (p=0.587)
- Forecast stabilizes at ~12.8 million dollars (based on test set trend).
- Confidence Intervals to be visualized.

```python
# Split data
train_size = int(len(df) * 0.8)
train, test = df['Revenue'][:train_size], df['Revenue'][train_size:]
print("\nTrain size:", len(train), "Test size:", len(test))
print("Train data sample (last 5):", train.tail())
print("Test data sample (first 5):", test.head())

# Fit ARIMA model
model = ARIMA(train, order=(1, 1, 1))
fit = model.fit()
print("\nARIMA(1,1,1) Summary:")
print(fit.summary())

# Forecast
forecast_steps = 90  # 3 months
forecast = fit.forecast(steps=forecast_steps)
conf_int = fit.get_forecast(steps=forecast_steps).conf_int()
forecast_index = pd.date_range(start=test.index[0], periods=forecast_steps, freq='D')
forecast_series = pd.Series(forecast, index=forecast_index)
print("\nForecast for next 90 days:")
print(forecast_series.head())
print("\nConfidence Intervals:")
print(conf_int.head())
```

## F1-F2: Results and Visualization
Evaluate MSE and visualize the forecast.

**Output**:
- MSE: 12.77
- Forecast plot shows training data increasing to ~24.8 million, test data declining from ~12.8 million, forecast stable at ~12.8 million with 95% CI.

```python
# Evaluate model
test_forecast = fit.forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast, index=test.index)
mse = ((test - test_forecast) ** 2).mean()
print(f"\nMean Squared Error: {mse:.2f}")

# Visualize forecast
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data')
plt.plot(forecast_series.index, forecast_series, label='Forecast', color='green')
plt.fill_between(forecast_series.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='green', alpha=0.3, label='95% Confidence Interval')
plt.title('ARIMA Forecast of Daily Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions of Dollars)')
plt.legend()
plt.grid(True)
plt.savefig('forecast_plot.png')
plt.show()
```
